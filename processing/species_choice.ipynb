```python
import pandas as pd

#pick only the strains used in the paper
import pandas as pd

all_strains_path = "../literature/all_strains_form_supplement.csv"
chosen_strains_path = "../literature/chosen_strains_from_supplement.csv"

df = pd.read_csv(all_strains_path, header=None, names=["Species", "Strain"])

# pick only lines ending with T
df_chosen = df[df['Strain'].str.strip().str.endswith('T')].copy()

# remove T from the end
df_chosen['Strain'] = df_chosen['Strain'].str.strip().str.rstrip('T')

df_chosen.to_csv(chosen_strains_path, index=False, header=False)
```

```python
import os
import pandas as pd

accessions_path = "accessions.csv"
data_dir = "../data"

# load accessions
df = pd.read_csv(accessions_path)

# check files in folder
files_in_data = os.listdir(data_dir)

# identify missing
missing = []
for _, row in df.iterrows():
    # reconstruct expected filename
    expected_name = row['Species'].replace(" ", "_").replace(".", "").replace(":", "") + ".faa"
    if expected_name not in files_in_data:
        missing.append(expected_name)

print(f"Missing files: {missing}")
```

Bifidobacterium saeculare is missing because it has been reclassified as a subspecies of Bifidobacterium gallinarum and was unlisted from the NCBI database.

Scardovia inopinata was addded manually because it's not present in the supplementary table from the paper.
